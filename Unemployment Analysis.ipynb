```python
df['Area'].value_counts().plot(kind='pie',labels=['Urban','Rural'], autopct='%1.2f%%',explode=[0.1,0])
plt.title('Distribution of Area')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(15,8))
df['Region'].value_counts().plot(kind='bar')
plt.xlabel('State')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize =  (20,12))
data_grouped = df.groupby('Region')[' Estimated Unemployment Rate (%)'].sum()
data_grouped.plot(kind='pie', autopct='%1.2f%%')
plt.title('Distribution of Estimated Unemployment Rate by Region')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize =  (20,12))
data_grouped = df.groupby('Region')[' Estimated Labour Participation Rate (%)'].sum()
data_grouped.plot(kind='pie', autopct='%1.2f%%')
plt.title('Distribution of Estimated Labour Participation Rate by Region')
plt.axis('equal')
plt.show()
```
